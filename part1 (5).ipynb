```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def combineForegroundBackground(fgImg, fgMask, bgImg, topLeft):
    # To take the dimensions of foreground and mask
    fgHeight, fgWidth, _ = fgImg.shape

    # To take the dimensions of the background
    bgHeight, bgWidth, _ = bgImg.shape

    # Calculate the bottom right corner of the foreground image in the background image
    fgBottomRight = (topLeft[0] + fgWidth, topLeft[1] + fgHeight)

    # For adjusting the foreground to the background 
    for y in range(topLeft[1], fgBottomRight[1]):
        for x in range(topLeft[0], fgBottomRight[0]):
            
            # We are checking if the foreground object is in the bounds of the background
            if x >= 0 and y >= 0 and x < bgWidth and y < bgHeight:
                # Get the foreground and mask to the corresponding pixels
                fgPixel = fgImg[y - topLeft[1], x - topLeft[0]]
                fgMaskPixel = fgMask[y - topLeft[1], x - topLeft[0]]

                # Check if the pixel belongs to the foreground object
                if fgMaskPixel > 0:
                    # Update the pixel in the background image with the pixel from the foreground image
                    bgImg[y, x] = fgPixel

    return bgImg

# Load the photos
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')


# Convert foreground and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

topLeft = (0, 0)

# Calling combineForegroundBackground function
result = combineForegroundBackground(fgImg, fgMask, bgImg, topLeft)

# Show the resulting image as a plot
plt.imshow(result)
plt.title("topLeft = (0,0)")
plt.show()


# Load foreground image, foreground mask, and background image
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')


# Convert foreground and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

# Call the combineForegroundBackground function with topLeft = (1000,1000)
topLeft = (1000,1000)
result1 = combineForegroundBackground(fgImg, fgMask, bgImg, topLeft)


# Load foreground image, foreground mask, and background image
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')


# Convert foreground image and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

# Call the combineForegroundBackground function with topLeft = (250,250)
topLeft = (250,250)
result2 = combineForegroundBackground(fgImg, fgMask, bgImg, topLeft)


# Load foreground image, foreground mask, and background image
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')


# Convert foreground image and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

# Call the combineForegroundBackground function with topLeft = (250,350)
topLeft = (250,350)
result3 = combineForegroundBackground(fgImg, fgMask, bgImg, topLeft)


# Load foreground image, foreground mask, and background image
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')


# Convert foreground image and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

# Call the combineForegroundBackground function with topLeft = (700,400)
topLeft = (700,400)
result4 = combineForegroundBackground(fgImg, fgMask, bgImg, topLeft)

# Show the resulting images as plots
fig, axs = plt.subplots(1, 4, figsize=(15, 5))
axs[0].imshow(result1)
axs[0].set_title("topLeft = (1000,1000)")
axs[1].imshow(result2)
axs[1].set_title("topLeft = (250,250)")
axs[2].imshow(result3)
axs[2].set_title("topLeft = (250,350)")
axs[3].imshow(result4)
axs[3].set_title("topLeft = (700,400)")
plt.show()


# Load foreground image, foreground mask, and background image
fgImg = cv2.imread('0431.jpg')
fgMask = cv2.imread('mask0431.png', 0)
bgImg = cv2.imread('mars.jpeg')

# Convert foreground and background image to RGB
fgImg = cv2.cvtColor(fgImg, cv2.COLOR_BGR2RGB)
bgImg = cv2.cvtColor(bgImg, cv2.COLOR_BGR2RGB)

# Show the resulting images as plots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(fgImg)
axs[0].set_title("Foreground Image")
axs[1].imshow(fgMask,cmap='gray')
axs[1].set_title("Foreground Mask")
axs[2].imshow(bgImg)
axs[2].set_title("Background Image")
plt.show()

```
